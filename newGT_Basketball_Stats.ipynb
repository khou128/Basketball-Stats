```python
import pandas as pd
```

```python
url_1 = 'https://raw.githubusercontent.com/khou128/Basketball-Stats/main/cleaned_basketball_nba.csv'
df_1 = pd.read_csv(url_1)
df_1
```

```python
from scipy.stats.mstats import winsorize

# Assuming df is your DataFrame and 'column_name' is the column containing the data with extreme values
# For example, let's say you want to winsorize the 'value' column
winsorized_values = winsorize(df_1['usg_pct'], limits=[0.05, 0.05])  # Winsorize values below the 5th percentile and above the 95th percentile

# Replace the original column with the winsorized values
df_1['usg_pct'] = winsorized_values
```

```python
#Mean of True Shooting Percentage:
column_name = 'ts_pct'
mean_value = df_1[column_name].mean()

print("Mean of column '{}' is: {}".format(column_name, mean_value))

```
